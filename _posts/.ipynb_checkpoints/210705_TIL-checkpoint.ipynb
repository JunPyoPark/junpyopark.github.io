## Reading

### Article
<strong>[Lessons from the pandemic’s superstar data scientist, Youyang Gu](https://www.technologyreview.com/2021/04/27/1023657/lessons-from-the-pandemics-superstar-data-scientist-youyang-gu/)</strong>

<ul>
<li>Lesson #1: Focus on fundamentals (심플한 모델, 데이터가 realistic 문제에 더 working 할 수 있다)
<li>Lesson #2: Minimize assumptions (선입견을 제거하고 데이터를 봐라)
<li>Lesson #3: Test the hypothesis (가설수립과 테스트는 unit action이다)
<li>Lesson #4: Learn from mistakes (예측실패에 대해 분석 후 보완해라)
<li>Lesson #5: Engage critics (비판적 사고와 토론으로 confirmation bias을 제거해라)
<li>Lesson #6: Exercise healthy skepticism (건전한 회의주의 사고를 가져라)
